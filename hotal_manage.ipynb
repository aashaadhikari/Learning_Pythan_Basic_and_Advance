```python
class room:
    def __init__(self, room_number, capacity, is_booked = False):
        # false is used Because to show room is not booked now
        self.room_number = room_number
        self.capacity = capacity
        self.is_booked = is_booked
```

```python
class Hotel:
    def __init__(self, name, num_room): # define a hotal with there attribute
        #self.name = name
        self.rooms = []
        for room_number in range(1, num_room + 1):
            capacity = room_number
            self.rooms.append(room(room_number, capacity))
        # self.rooms = [room(room_number, capacity) for room_number, capacity in enumerate(range(1, num_room+1), start = 1)]
    

    def display_available_room(self): # defining method display all room
        available_rooms = [room for room in self.rooms if not room.is_booked]

        if available_rooms:
            print("Available Rooms: ")
            for room in available_rooms:
                print(f"Room Number: {room.room_number}, capacity: {room.capacity}")

        else:
            print("no Rooms available.")

    def book_room(self, room_number): # defining method to book room
        room = self.rooms[room_number - 1]

        if not room.is_booked:
            room.is_booked = True
            print(f"Room {room_number} booked successfully.")

        else:
            print(f"Room {room_number} is alrady booked.")

    def chackout_room(self, room_number):
        room = self.rooms[room_number - 1]
        if room.is_booked:
            room.is_booked = False
            print(f"Room {room_number} checked out sucessfully. ")

        else:
            print(f" Room {room_number} is noot booked. ")

```

```python
def main():
    hotel = Hotel("Example Hotel", 10)

    while True:
        print("\n Hotal management system")
        print("1. Display Avilable rooms")
        print("2. book a Room")
        print("3. Check out of a Room")
        print("4. exit")

        choice = input("Enter the number what you want: ")
        if choice == "1":
            hotel.display_available_room()

        elif choice == "2":
            room_number = int(input("Enter the room number to booked: "))

            if 1 <= room_number <= len(hotel.rooms):
                hotel.book_room(room_number)

            else:
                print("Invalid Room Number")

        elif choice == "3":
            room_number = int(input("Enter the room number to chacked out: "))

            if 1 <= room_number <= len(hotel.rooms):
                hotel.chackout_room(room_number)

            else:
                print("Invalid room number.")

        elif choice == "4":
            print("Thanks for using this hotel and good)bye")
            break # tells python to stop the game

        else:
            print("Invalid choice. Please enter a valid option")

if __name__ == "__main__":
    main()
        
```
