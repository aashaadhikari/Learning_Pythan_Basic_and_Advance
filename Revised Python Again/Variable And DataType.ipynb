# Valid Identifier Rules:
* Can contain letters (A–Z, a–z), digits (0–9), and underscores (_)

    * Example: my_var, data1, user_name
* Must begin with a letter or an underscore (_)

    * Valid: _temp, name
    * Invalid: 1name (starts with a digit)

* Cannot use Python keywords or reserved words

    * Example (invalid): for, if, class, def, return, etc.

* Identifiers are case-sensitive

    * Name, name, and NAME are considered different identifiers.

* Cannot contain special characters

    * Invalid: name$, user-name, data@home

* Should not start with a digit

    * Invalid: 2value, 9_data
 
#  Best Practices (Not strict rules):
* Use meaningful names: student_name is better than sn.
* Use snake_case for variables and functions: total_price, get_data()
* Use PascalCase for class names: StudentDetails, EmployeeRecord
* Avoid using single letters like l (lowercase L) or O (capital O) as they can be confused with digits 1 and 0.
