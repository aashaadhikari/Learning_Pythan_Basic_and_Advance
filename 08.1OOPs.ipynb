# Python OOPs Concepts
Python OOPs, we'll explore the core principles of object-oriented programming (OOP) in Python. From encapsulation to inheritance, polymorphism, abstract classes and iterators.

**OOP (Object-Oriented Programming)** is a programming style based on the concept of **"objects"**, which contain **data** (attributes) and **behavior** (methods).
It helps to organize code, make it reusable, and easier to manage.

### Key Concepts of OOP:

* **Class**: A blueprint for creating objects.
* **Object**: An instance of a class.
* **Encapsulation**: Hiding internal details and showing only what’s needed.
* **Inheritance**: A class can inherit features from another class.
* **Polymorphism**: Same function name can have different meanings in different contexts.
* **Abstraction**: Hiding complex details and showing only essentials.

### Why to use OOP
* Reusability
* modularity
* maintainability
* scalability
* interesting Tidbit

# Note:
Python is **not considered fully object-oriented** because it supports both **object-oriented** and **procedural (functional)** programming styles.

### Reasons why Python is *not fully* OOP:

1. **Primitive data types** like `int`, `float`, and `bool` can be used without creating objects, even though internally they are classes.
2. You can write programs in Python without using **classes or objects**—using just functions and variables (procedural style).
3. Python allows mixing **object-oriented** and **non-OOP** features freely.

### But…

* Internally, **everything in Python is an object**, even functions and types.
* Python supports **OOP features** like encapsulation, inheritance, and polymorphism.

So, while Python is *strongly* object-oriented, it's **not strictly OOP** like languages such as Java or Smalltalk.
