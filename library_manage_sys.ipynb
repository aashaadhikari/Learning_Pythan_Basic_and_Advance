```python
class Library:
    def __init__(self):
        self.books = []
    def add_book(self, book_id, title, author, collection):
        self.books.append([book_id, title, author,collection])
        return f"The given book is added sucessfully"

    def display_avia_book(self):
        for i in range(len(self.books)):
            if self.books[i][3] > 0:
            # if self.books[i][3] !=0 and self.books[i][3] >0:
                print(f"{self.books[i][0]}-{self.books[i][1]}-{self.books[i][2]}-{self.books[i][3]}")
           
    def borrow_book(self, title):
        for i in range(len(self.books)):
            if self.books[i][1] == title:
                if self.books[i][3] > 0:
                    print("You  borrowed the book")
                    self.books[i][3] -= 1
                else: 
                    print("Book is not available right now.")


    def book_return(self, title):
        for i in range(len(self.books)):
            if self.books[i][1] == title:
                self.books[i][3] += 1
                print("the borrowed book is reatern")
           
   
           
```
