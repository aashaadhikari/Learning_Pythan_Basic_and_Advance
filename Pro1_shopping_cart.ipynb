```python
print("This is my shopping cart project")

foods = []
prices = []
total = 0

while True:
    food = input("Enter the food you want to order(q for exit)")
    if food.lower()=="q":
        break 

    else:
        price = float(input(f"Enter the price of food:{food}:$"))
        foods.append(food)
        prices.append(price)

print("-----------your cart-----------")

for food in foods:
    print(food)

for price in prices:
    total +=price

print()
print(f"The Total price is: {total}")
```
